```
import pandas as pd
from fbprophet import Prophet
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
import numpy as np
```

```
robo= pd.read_csv('https://raw.githubusercontent.com/alvinwalker314/DS-Unit-2-Regression-1/master/module1-making-forecasts/pageviews-20150701-20190523.csv')
```

```
robo.rename(columns= {'Robotics': 'y', 'Date': 'ds'}, inplace= True)
```

```
robo= robo.set_index('ds')
```

```
train= robo['2016':'2018']
test= robo['2018':'2019']
```

```
train= train.reset_index()
test= test.reset_index()
```

```
ax= train.plot(x='ds',y= 'y', figsize= (10,5))
test.plot(x='ds', y='y', color='orange', ax=ax)
```

```
train['mean_baseline']= train['y'].mean()
test['mean_baseline']= train['y'].mean()
```

```
ax=train.plot(x='ds', y='y', figsize= (10, 5))
train.plot(x='ds', y= 'mean_baseline', ax= ax)
```

```
train_meanbl_mae= mean_absolute_error(test['y'], test['mean_baseline'])
test_meanbl_mae= mean_absolute_error(test['y'], test['mean_baseline'])
```

```
print('Train Mean Absolute Error(meanbaseline): ',train_meanbl_mae)
print('Test Mean Absolute Error(meanbaseline): ', test_meanbl_mae)
```

```
ax= test.plot(x='ds', y='mean_baseline', figsize= (10, 5))
test.plot(x='ds', y='y', ax=ax)
```

```
train['naive_baseline']= train['y'].shift(1).bfill()
test['naive_baseline']= test['y'].shift(1).bfill()
```

```
ax=train.plot(x='ds', y='naive_baseline', figsize= (10,5))
train.plot(x='ds', y='y', color='orange', ax=ax)
```

```
ax= test.plot(x='ds', y='naive_baseline', figsize=(10,5))
test.plot(x= 'ds', y='y', ax=ax)
```

```
train_naive_mae= mean_absolute_error(train['y'], train['naive_baseline'])
test_naive_mae= mean_absolute_error(test['y'], test['naive_baseline'])
```

```
print('Train Mean Absolute Error(naive baseline): ', train_naive_mae)
print('Test Mean Absolute Error(naive baseline): ', test_naive_mae)
```

```
model= Prophet(daily_seasonality= False)
model.fit(train)
train_predictions= model.predict(train)
test_predictions= model.predict(test)
```

```
ax= train_predictions.plot(x= 'ds', y='yhat')
```
